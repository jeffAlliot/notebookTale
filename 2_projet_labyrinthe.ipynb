```python
# =============================================================================
# MINI PROJET LABYRINTHE
# =============================================================================
import matplotlib.pyplot as plt
from random import randint
import os
import imageio

#! pip install ???
            
class File():
    """création de la classe File avec le type list"""
    def __init__(self):
        """initialisation File vide"""
        self.liste=[]
        
    def vide(self):
        """teste si la File est vide"""
        if self.liste==[] :
            return True
        else :
            return False
        
    def defiler(self):
        if self.vide()==False :
            return self.liste.pop(0)  #1er élément donc le 1er entré
    
    def enfiler(self,x):
        self.liste.append(x)
        
    def affichage(self):
        for i in range(len(self.liste)) : 
            print(self.liste[-i-1],end=" ")
        print("")
        
    def sommet(self):
        """renvoie le sommet de la File"""
        return self.liste[0]

#fonctions

def voisin(G:dict,sommet:str)->list:
    """retourne les voisins d'un sommet"""
    return(G[sommet])

def trouve_chemin(G:dict,som_deb:str,som_fin:str,chaine:list)->list:
    """retourne UN chemin entre som_deb
    et som_fin
    au départ chaine = []"""
    chaine.append(som_deb)
    if som_deb==som_fin :
        return(chaine)
    for elt in voisin(G,som_deb):
        if elt not in chaine :
            nouveau_chemin=chemin(G,elt,som_fin,chaine)
            if nouveau_chemin!=None:
                return nouveau_chemin
    return(None)
    
      
    

def dict_largeur(G:dict,som_deb:str)->dict :
    """retourne un dictionnaire qui
    contient les sommets visités en largeur et
    leur parent"""
    parent = {}
    visites = []
    f = File()
    f.enfiler(som_deb)
    parent[som_deb]=None
    while  f.vide()==False:
        tmp=f.defiler()
        if tmp not in visites:
            visites.append(tmp)
        for elt in voisin(G,tmp):
            if elt not in visites and elt not in f.liste:
                f.enfiler(elt)
                parent[elt] = tmp
    return(parent)   
    

def solution(end:str,parents:dict)->list:
    """retourne le chemin en remontant les
    parents à partir de end """
    chemin =[]
    courant = end
    while courant != None :
        chemin = [courant] + chemin
        courant = parents[courant]
    return(chemin)

def chemin_court(G:dict,som_deb:str,som_fin:str)->list:
    """retourne le chemin le plus court entre
    som_deb et som_fin grâce à un parcours en largeur
    le graphe est NON pondéré """
    parent = dict_largeur(G,som_deb)
    return (solution(som_fin,parent))
```
